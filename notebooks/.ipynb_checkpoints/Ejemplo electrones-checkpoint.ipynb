# Ejemplo para usar LegPy con electrones

Importamos las librerías y definimos el haz de electrones

```python
import LegPy as lpy
import numpy as np
import matplotlib.pyplot as plt

E = 1. # MeV
beam = lpy.Beam(name='parallel', theta=0.0, phi=0.0, particle="electron")
spectrum = lpy.Spectrum(name='mono', E=E)
```

Definimos el medio y el tamaño de paso tomando 1% del CSDA

```python
medium = lpy.Medium(name='Al') # NIST
e_data = medium.e_data
CSDA = np.interp(E, e_data.E_ref, e_data.R_ref) # cm
CSDAum = CSDA * 1.e4 # um
e_length = CSDAum * 0.01 # um
print('CSDA = ', round(CSDA, 3), 'cm')
```

Valor de f por defecto

```python
X0 = e_data.X0
f = 0.11803 / X0**0.5 + 1.4142
print(f)
```

Opcional: comprobamos la validez de la aproximación gaussiana para el tamaño de paso y el valor de f elegido

```python
e_data.make_step_list(0.01, e_length, None)#, f=1.454, g=0.)
mean_scat_degrees = np.degrees(e_data.mean_scat) # degrees
E_min = e_data.E[mean_scat_degrees<10.].min()
R_tail = e_data.R[mean_scat_degrees<10].min()
print("La aproximación gaussiana es válida (mean_scat < 10 deg.) para E >", round(E_min, 2), 'MeV')
print("A esta energía, al electrón le queda por recorrer", round(R_tail,2), "cm", "=", round(100.*(1.-R_tail/CSDA),1), "%CSDA")
```

Definimos la geometría del medio conforme al CSDA y ejecutamos la simulación con el tamaño de paso y el valor de f elegidos

```python
geometry = lpy.Geometry(name='cylinder', z=0.4, r=0.5, n_z=10, n_r=10) # cm
hist, E_dep, flu = lpy.MC(medium, geometry, spectrum, beam,
                          n_part = 1000,
                          histograms=False,
                          n_z=50,
                          n_ang=30,
                          Edep_matrix=True,
                          e_length=e_length,
                          #e_f=1.454,
                         )
```

En el DataFrame hist se guardan las distribuciones de z final (directa y acumulada), z máxima (directa y acumulada), y ángulo theta de electrones rebotados (directa y normalizada por ángulo sólido)

```python
#hist[0].head()
#hist[0].plot(kind='scatter', x=0, y=1);
hist[0].plot(kind='scatter', x=0, y=2);
#hist[1].head()
#hist[1].plot(kind='scatter', x=0, y=1);
#hist[1].plot(kind='scatter', x=0, y=2);
#hist[2].head()
#hist[2].plot(kind='scatter', x=0, y=1);
#hist[2].plot(kind='scatter', x=0, y=2);
```

Calculamos el rango extrapolado a partir de la distribución de z final o del z máximo

```python
R_ext, R_max, R_av = lpy.ext_range(hist[0])
#R_ext, R_max, R_av = lpy.ext_range(hist[1])
```
